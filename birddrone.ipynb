```python
from ultralytics import YOLO

# Load a YOLOv8 model (you can also use yolov8n/yolov8m/yolov8l/yolov8x)
model = YOLO('yolov8n.pt')  # Using the small 'nano' model for fast training

# Train the model
model.train(data='/content/Drone_Bird-1/data.yaml', epochs=50, imgsz=640)
```
